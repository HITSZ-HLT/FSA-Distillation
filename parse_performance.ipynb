```python
import json

# Replace with your actual path
path = "./evaluation/output/2025-06-12/performance.txt" 
data = []

with open(path, mode='r', encoding='utf-8-sig') as f:
    for line in f:
        data.append(json.loads(line))

# Replace with the version you want to filter by
version = 'v7.9' 

data_version = [dict for dict in data if version in dict['model_name_or_path'] and 'atsa' in dict['dataset']]

sorted_list = sorted(data_version, key=lambda x: x['seed'])

# 种子列表
seeds = [42, 52, 62, 72, 82, 142, 152, 162, 172, 182]

# 添加表头
print("TSA-Rest14, TSA-Laptop14, F1 Score Results (version: {})".format(version))
print("=" * 60)
print("Seed-based F1 Scores (out of 100.00):")
print("-" * 60)

# 假设数据按种子和数据集组织，每个种子有多个数据集结果
# 重新组织数据结构
results_by_seed = {}
datasets_per_seed = len(sorted_list) // len(seeds) if len(sorted_list) >= len(seeds) else 1

# 将数据按种子分组
for i, result in enumerate(sorted_list):
    seed_index = i // datasets_per_seed
    if seed_index < len(seeds):
        seed = seeds[seed_index]
        if seed not in results_by_seed:
            results_by_seed[seed] = []
        results_by_seed[seed].append(result['metric']['f1'] * 100)


# 输出每个种子的结果
all_columns = []
for seed in seeds:
    if seed in results_by_seed:
        print(f"{seed:<6}", end='')
        for j, score in enumerate(results_by_seed[seed]):
            print(f"{score:8.2f}", end='')
            # 收集每列数据用于计算平均值
            if len(all_columns) <= j:
                all_columns.append([])
            all_columns[j].append(score)
        print()
    else:
        # 如果某个种子没有数据，可以跳过或显示N/A
        print(f"{seed:<6}", end='')
        for j in range(datasets_per_seed):
            print(f"{'N/A':>8}", end='')
        print()

# 计算并输出平均值
print("-" * 60)
print(f"{'Avg':<6}", end='')
for col_data in all_columns:
    if col_data:
        avg = sum(col_data) / len(col_data)
        print(f"{avg:8.2f}", end='')
    else:
        print(f"{'N/A':>8}", end='')
print()
print("-" * 60)
```
